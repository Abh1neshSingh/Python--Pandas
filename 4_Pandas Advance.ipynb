```python
df1=pd.DataFrame({'key1':[1,2,4,5,6],
              'key2':[4,5,6,7,8],
              'key3':[3,4,5,6,6]
})
```

```python
df1
```

```python
df2=pd.DataFrame({'key1':[1,2,45,5,67],
              'key4':[44,5,36,7,84],
              'key5':[3,44,5,46,6]
})
```

```python
df2
```

```python
# first see common columns 
```

```python
# merge
# merge join based upon the columns
```

```python
pd.merge(df1,df2,how='left')
```

```python
pd.merge(df1,df2,how='right')
```

```python
pd.merge(df1,df2,how='outer') # all the keys of both the dataframe
```

```python
pd.merge(df1,df2,how='outer')
```

```python
pd.merge(df1,df2,how='inner')
```

```python
# cross join
```

```python
pd.merge(df1,df2,how='cross')
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,how='left',left_on='key2',right_on='key4')
```
